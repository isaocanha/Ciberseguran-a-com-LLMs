# **Resumo (*Abstract*)**

Com o aumento significativo dos ataques cibernéticos nos últimos anos, a segurança da informação tornou-se uma preocupação central para diversas organizações. Estimativas indicam que, em 2021, os danos causados por ciberataques alcançaram U\$6 trilhões globalmente, e projeta-se que esse número atinja U\$10,5 trilhões até 2025. Neste contexto, o uso de técnicas avançadas para a detecção e mitigação de vulnerabilidades em software é essencial. Este trabalho tem como objetivo desenvolver um modelo de Large Language Model (LLM) para identificar vulnerabilidades em código estático e sugerir melhorias, promovendo um desenvolvimento seguro. A pesquisa envolve a análise de soluções existentes, a identificação de tipos de vulnerabilidades, a criação de datasets de código vulnerável e não vulnerável, e o fine-tuning de um modelo de LLM genérico. A hipótese é que LLMs treinados para analisar código-fonte são eficazes na identificação de vulnerabilidades, auxiliando na redução de falhas não detectadas e contribuindo para o desenvolvimento de aplicações mais seguras. A metodologia inclui uma prova de conceito com um modelo pré-treinado, seguida pelo fine-tuning com datasets especializados, avaliação iterativa e refinamento do modelo. Espera-se que LLMs se provem eficientes na cibersegurança melhorando a detecção de vulnerabilidades em código estático.

# **Referencial Teórico**

A segurança da informação é responsável por garantir que os dados de empresas
sejam confidenciais, confiáveis e acessíveis, reduzindo, assim, o risco de ataques cibernéticos
(BARRETO et al., 2018). Portanto, é necessário que a empresa consiga minimizar os
potenciais ataques de segurança, desse jeito, pode-se considerar a Inteligência Artificial
(IA) como uma ótima aliada a segurança da informação, já que pode desempenhar um
papel crucial na detecção precoce de vulnerabilidades em códigos.


A IA surgiu na área da ciência da computação, visando capacitar máquinas para
que fossem aptas a realizar atividades semelhantes às humanas, logo, podendo ser sistemas
que pensam ou agem como seres humanos, ou sistemas que pensam ou agem racionalmente.
Dessa maneira, a IA possui uma ampla variedade de aplicações, assim tendo subcampos
como: resolução de problemas, processamento de linguagem natural, aprendizagem de
máquina, rede neural artificial, entre outros (VALDATI, 2020).

O aprendizado de máquina (machine learning), por exemplo, é uma técnica baseada
na criação de modelos que são determinados por meio de uma coleção de dados ou
reconhecendo padrões. Já as Redes Neurais Artificiais (RNA), são modelos computacionais
que tentam imitar a rede neural do cérebro humano. Uma das estruturas de RNA mais
básicas é o Perceptron, que é baseado em uma única camada de unidade lógica de limiar
(threshold logic unit (TLU)), sendo assim, apenas um neurônio artificial. Entretanto,
problemas complexos não podem ser resolvidos por simplesmente uma arquitetura de rede
neural com poucos Perceptrons. Para resolver problemas mais complexos, são utilizadas
técnicas de aprendizagem profunda (deep learning), em que várias camadas são definidas
na arquitetura da rede (FERREIRA, 2021).

Outro ramo da IA, que se utiliza fortemente de deep learing, é o processamento
de linguagem natural, que ajuda as máquinas a processarem, analisarem e responderem
dados de linguagem natural. Dentro deste campo aparece uma nova vertente, os grandes
modelos de linguagem (large language model (LLM)), que se refere a modelos treinados
com um grande volume de dados e muitos parâmetros. Um exemplo de LLM é o Generative
Pre-trained Transformer (GPT), que é um modelo autoregressivo, ou seja, gera uma
palavra de cada vez e usa palavras geradas anteriormente como contexto para palavras
futuras. Geralmente o GPT tem que passar por fine-tuning, que é o processo de adaptar
um LLM pré-treinado para uma tarefa específica, adicionando camadas de revisão humana
ou desenvolvendo sistemas de verificação externos (AMARATUNGA, 2023).

Das aplicações gerais dos LLMs, foi escolhido estudar a detecção de vulnerabilidades
e segurança que é uma área promissora no mercado. Para realizar a definição das técnicas

tema. Desse modo, com o auxílio das bases de dados IEEE Xplorer e Elsevier foram
encontrados artigos utilizando duas palavras-chave: Large Language Models e Vulnerability
detection in codes using LLM, com esse filtro foram coletados seis artigos, porém, acabamos
foram selecionados quatro que se mostraram fundamentais para a elaboração do projeto.

De acordo com Yao et al. (2024) a detecção e correção de vulnerabilidades em
códigos utilizando LLMs tem um impacto muito positivo na cibersegurança. Tendo em vista
um estudo citado feito com LLMs, esses foram capazes de identificar, significativamente
mais vulnerabilidades no código em comparação as ferramentas mais tradicionais de análise
estática em código, entretanto, ainda houve uma grande ocorrência de falsos positivos,
impactando a eficácia deles. Por isso, conforme Raiaan et al. (2024), para alcançar resultados
satisfatórios deve-se combinar um LLM a um ótimo dataset e fine-tuning. Por exemplo,
os modelos de LLMs GPT, que oferecem para os desenvolvedores uma boa capacidade de
aproveitar uma modelagem avançada de linguagens para uma variedade diversificada de
aplicações, como o desenvolvimento seguro.

Logo, segundo Li e Shan (2023) LLMs treinados são capazes de varrer e filtrar
vulnerabilidades em códigos, além de analisarem as interfaces de programação de aplicação
(application programming interface (API)) e criar regras para proteger códigos. Em seu
trabalho foram utilizadas duas LLMs bases: GPT 3.5 e GPT 4, e duas técnicas de fine-
tuning: Low-Rank Adaptation (LoRa) e Query-LoRa (QLoRa). Como o objetivo era
detectar potenciais vulnerabilidades lógicas, os LLMs poderiam ser treinados com os
seguintes propositos: reconhecimento de padrões, análise de vulnerabilidades, compreensão
de linguagem, interferência e conhecimento geral. Após seguir todos os passos necessários
para aplicar os LLMs na área de cibersegurança, Li e Shan (2023) concluíram que a análise
feita usando GPT 3.5 obteve resultados diferentes do GPT 4, o primeiro forneceu como
solução para o usuário as potenciais ameaças presente no código para que este conseguisse
reduzir suas vulnerabilidades, já o segundo sugeriu possíveis resoluções para o usuário ter
como base o que poderia ser ajustado para obter um código seguro.

Foram avaliados quatro LLMs por Purba et al. (2023) para determinar se estes eram
capazes de classificar um determinado trecho de código como contendo ou não contendo
uma determinada vulnerabilidade de software. Para isso selecionou-se dois tipos de datasets:
code gadgets, e CVEfixes. Porém para o segundo foi mencionado que havia problemas de
qualidade de dados, dessa forma, os resultados obtidos tiveram um desempenho inferior.
Os LLMs escolhidos foram: GPT 3.5-Turbo e GPT 3.5-Turbo-0613 para comandos diretos,
e Davinci e codegen2B-multi para fine-tuning. Os resultados alcançados para detecção de
vulnerabilidades, após fazer testes com vulnerabilidades do tipo overflow e SQL Injection,
com os modelos GPT 3.5-Turbo e GPT 3.5-Turbo-0613 foram péssimos, já para os modelos
Davinci e codegen2B-multi os resultados foram um pouco melhor, apesar da precisão ser
ruim, contendo uma alta taxa de falsos positivos. Portanto, da análise desse trabalho, foi
identificado que o uso de LLMs pode se mostrar promissor para encontrar padrões de
vulnerabilidade; no entanto, o uso desses modelos deve ser combinado com outras técnicas
para que sua precisão seja adequada na aplicação de problemas reais

````python
from openai import OpenAI
client = OpenAI()

with open('codigo.py', 'r') as file:
  codigo = file.read()

completion = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {"role": "system", "content": "Voce eh um assistente IA perito em analise de seguranca de codigos alem de identificar vulnerabilidades e sugerir correcoes."},
    {"role": "user", "content": f"Existe alguma vulnerabilidade de seguranca no seguinte codigo? se sim de o codigo corrigindo as vulnerabilidades:\n\n{codigo}"}
  ]
)

resposta = completion.choices[0].message.content
print(resposta)
````

Neste exemplo o arquivo codigo.py continha o seguinte código vulnerável, contendo a vulnerabilidade de SQL Injection:

````python
@app.route("/login")
def login():

  username = request.values.get('username')
  password = request.values.get('password')

  db = pymysql.connect("localhost")
  cursor = db.cursor()

  cursor.execute("SELECT * FROM users WHERE username = '%s' AND password = '%s'" % (username, password))

  record = cursor.fetchone()
  if record:
    session['logged_user'] = username

  db.close()
````

Então ao executarmos o primeiro código com a entrada a baixo, obtemos a seguinte saída:

*Sim, este código é vulnerável a ataques de SQL Injection, pois utiliza diretamente os valores de `username` e `password` na query SQL, o que permite que um atacante execute comandos maliciosos no banco de dados.*

*A forma correta de proteger o código contra SQL Injection é utilizar prepared statements com parâmetros. Aqui está o código corrigido:*

```python
from flask import request, session
import pymysql

@app.route("/login", methods=['POST'])
def login():
    username = request.form.get('username')
    password = request.form.get('password')

    # Prepare database connection
    db = pymysql.connect("localhost")
    cursor = db.cursor()

    # Execute SQL query with parameters
    query = "SELECT * FROM users WHERE username = %s AND password = %s"
    cursor.execute(query, (username, password))

    # If the query returns any matching record, consider the current user logged in.
    record = cursor.fetchone()
    if record:
        session['logged_user'] = username

    # Disconnect from server
    db.close()
```

*Neste código corrigido, os valores de `username` e `password` são passados como parâmetros na execução da query, o que previne a possibilidade de SQL Injection.*

Ou seja, o modelo identificou corretamente que havia uma vulnerabilidade de SQL Injection no exemplo e ainda sugeriu a correção do código como pedimos.

# **Uma ideia de Projeto de Aplicação**

Uma ideia de projeto seria integrar os LLMs treinados para analisar códigos-fonte em uma esteira de desenvolvimento de empresas. Esses modelos poderiam ser empregados automaticamente para revisar o código submetido pelos desenvolvedores antes de serem adicionados ao repositório principal. A implementação poderia funcionar da seguinte maneira, sempre que um códgio novo é adicionado ao repositório, ele automaticamente é enviado para o programa demonstrado, então a LLM analisa o código, gerando uma saida que sobrescreveria o arquivo original, um novo código com correções e sujestões aplicadas. Assim, permitindo a detecção precoce de vulnerabilidades de segurança e possíveis falhas, economizando tempo e dinheiro. Os resultados da análise do modelo poderiam ser integrados diretamente nos fluxos de trabalho da equipe de desenvolvimento, logo, fornecendo feedback imediato e sugerindo correções quando necessário. Desse modo, melhorando a segurança geral do softwares desenvolvidos pelas empresas.

Já existem diversas outras soluções que analisam códigos estáticos assim que submetidos para detecção de vulnerabilidade, conhecidos como SAST (Static Application Security Testing). A grande vantagem da abordagem de utilizar LLMs neste caso é que ela pode automatizar grande parte das correções de vulnerabilidades comuns, o que poupa muito tempo e esforço do desenvolvedor, além de resolver vulnerabilidades que poderiam passar esquecidas por serem aparentementes irrelevantes ou de baixa revelancia.
